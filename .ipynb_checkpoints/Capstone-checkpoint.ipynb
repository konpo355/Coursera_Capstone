### Scraping

First we import the BeautifulSoup module

```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
```

We use request to scrape the webside in a text format

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
data  = requests.get(url).text # content of the webpage in text format and stored in a variable called data
soup = BeautifulSoup(data,"html5lib")
```

In order to creat the data base we find the table in the soup and we initialize the dataframe. 

Then we go through all the rows of the table and assign the values to the appropreate column of the dataframe. 

Be careful since it has to become text and also we have to strip the \n from the end of each value.

```python
table = soup.find('table') #find a html table in the web page
df=pd.DataFrame(columns=["PostalCode", "Borough","Neighborhood"])

for row in table.find_all('tr')[1:]:
    # Get all columns in each row.
    cols = row.find_all('td') 
    # Store all information of the columns
    PostalCode = cols[0].string.rstrip('\n')
    Borough = cols[1].string.rstrip('\n')
    Neighborhood=cols[2].string.rstrip('\n')
    df=df.append({"PostalCode":PostalCode, "Borough":Borough,"Neighborhood":Neighborhood},ignore_index=True)
```

### Data Cleaning

Firstly we will take away all the rows with assigned borough. 


```python
# Drop all the not assigned borough by making them NaN and then droping that
df["Borough"]=df["Borough"].replace("Not assigned",np.nan)
df.dropna(axis='index',inplace=True)
```

In the case where there is a non assigned neighborhood left we assign it the Borough

```python
df['Neighborhood'] = np.where(df['Neighborhood'] == "Not assigned", df['Borough'], df['Neighborhood'])
```

We group by all the neighborhoods with the same Postal Code

```python
#We group all the neighborhoods by separating them with a ","
df=df.groupby("PostalCode", sort=False,as_index=False).agg( ','.join)
df.head()
```

After all the cleaning we reset the index of the dataframe

```python
# reset the index
df.reset_index(drop=True,inplace=True)
```

We find the shape of the cleaned dataframe.

```python
df.shape
```

### Exploring the Neighborhoods in Toronto

Let us create a function that finds all the venues in a 500meters radius from each neighborhood in Toronto.

#### GetNearbyVenues function

We are going to creat a function that will find all the venues around an input neighborhood.

To do that we will first initiate the list for the venues and then we will make a loop that goes through each of the neighborhoods. 

Inside the loop we will make the API call with a GET request in order to have access to all the venues.

Using that we will save only hthe relevant information which are the name, coordinates, and type of venue, and append them into the initiated list. 

Finaly from that list (after the loop has ended) we will create a new dataframe which has the neighborhood data from before plus the new information about the venue.

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]#initiate the list
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)#just to see the names of the venue
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    #create the dataframe 
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category'] #label the columns of the new dataframe
    
    return(nearby_venues)
```

#### 5 most common venues for each neighborhood

In order to do that we first write a function that sorts the venues in deschending order

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

Then we cread a new dataframe to display the top 10 venues for each neighborhood

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood Name']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood Name'] = tor_grouped['Neighborhood Name']

for ind in np.arange(tor_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(tor_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

### Clustering

We will cluster the neighborhoods using k-means. 

```python
# set number of clusters
kclusters = 3

tor_grouped_clustering = tor_grouped.drop('Neighborhood Name', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters).fit(tor_grouped_clustering)


# check cluster labels generated for each row in the dataframe
kmeans.labels_
```

Let's create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

tor_merged = tor_df

#merge manhattan_grouped with manhattan_data to add latitude/longitude for each neighborhood
tor_merged = tor_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood Name'), on='Neighborhood')

tor_merged.head() # check the last columns!
```

Finally, let's visualize the resulting clusters


```python
# create map
map_clusters = folium.Map(location=[43.6532, -79.3832], zoom_start=12)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(tor_merged['Latitude'], tor_merged['Longitude'], tor_merged['Neighborhood'], tor_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
